```python
#drop rows with NaN values (axis = 0 means row; 1 means column)
df2 = df.dropna(axis=0)
df2
```

```python
#fill numeric data with mean value
df3 = df.fillna(df.mean(numeric_only=True))
df3
```

```python
#fill NaN value (not numeric) with mode value
df4 = df3.fillna(df.mode().iloc[0])
df4
```
